```python
import pandas as pd
from datetime import date 
import calendar
import matplotlib.pyplot as plt
from sorted_months_weekdays import Month_Sorted_Month
%matplotlib inline
```

```python
df=pd.read_csv("Resources/individual_gifts.csv", encoding = "ISO-8859-1", parse_dates = ['Gift Date'])
df.head()
```

```python
df['year_of_gift'] = df['Gift Date'].map(lambda x: x.strftime('%Y'))
df['month_of_gift'] = df['Gift Date'].map(lambda x: x.strftime('%m'))
# df["month_of_gift"]=pd.to_numeric(df["month_of_gift"])
df["month_of_gift"] = df["month_of_gift"].astype(str).astype(int)
df['Gift Amount'] = df['Gift Amount'].str.replace(',', '').str.replace('$', '').astype(float)
df.head()
```

```python
df = df.sort_values(by='month_of_gift')
df.head()
```

```python
df = df[["Gift Amount", "year_of_gift", "month_of_gift"]]
df.head()
```

```python
ind_2014=df.loc[df['year_of_gift'] == "2014"]
ind_2014 = ind_2014.groupby(['month_of_gift']).agg({'Gift Amount': 'sum'})
ind_2014=ind_2014.reset_index()
ind_2014['month_of_gift'] = ind_2014['month_of_gift'].apply(lambda x: calendar.month_abbr[x])
ind_2014
```
